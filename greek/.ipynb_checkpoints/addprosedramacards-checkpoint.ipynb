```python
import re
basecards = {}
f = open('/Users/gcrane/github/canonical-greekLit/data/tlg0011/tlg002/tlg0011.tlg002.perseus-grc2.xml')
for l in f:
    m = re.search('<milestone unit="card" n="([^"]+)',l)
    if(m):
        basecards[m[1]] = 1

newcards = {}

def additem(curdict,curitem):
    if(curitem in curdict):
        curdict[curitem] = curdict[curitem] + 1
    else:
        curdict[curitem] = 1
        

for foo in card_alignment:
    additem(newcards,card_alignment[foo])

for foo in newcards:
    if(newcards[foo] > 1):
        print('extra',foo,newcards[foo])
    if(not foo in basecards):
        print('extra',foo)

for foo in basecards:
    if(not foo in newcards):
        print('missing',foo)
    
print(len(card_alignment))
```
