```python
import re
from lxml import etree

f = open("../translations/goodlines.tlg0011.tlg002.boeckh1884-deu1.xml")
tmpfname = '/tmp/tmp.xml'
fout = open(tmpfname,'w')
lnum = 0
for l in f:
    if(re.search('<l>',l)):
        lnum = lnum + 1
        l = re.sub('<l>','<l n="'+str(lnum)+'">',l)
    
    print(l,end='',file=fout)
fout.close()
f.close()
# 1. Parse your Boeckh text file
parser = etree.XMLParser(remove_blank_text=False)
tree = etree.parse(tmpfname, parser)
root = tree.getroot()
outfname = '../staging/tlg0011.tlg002.boeckh1884-deu2.xml'

namespaces = {'tei': 'http://www.tei-c.org/ns/1.0'}

def normalize_text(text):
    """Normalize Greek text for strict string comparison."""
    if not text:
        return ""
    text = text.replace("’", "'").replace("ʼ", "'")
    # Remove punctuation, spaces, and trailing dashes/hyphens
    text = re.sub(r'[^\w]', '', text)
    return text

# Gather all line elements from the XML
xml_lines = root.xpath("//tei:body//tei:l", namespaces=namespaces)

# 2. First Pass: Validate One-to-One Mapping
# We map each card query to the specific line elements it matches
match_tracker = {card_str: [] for card_str in card_alignment.keys()}

for line in xml_lines:
    line_raw_text = "".join(line.itertext())
    normalized_line = normalize_text(line_raw_text)
    
    for card_str in card_alignment.keys():
        if normalized_line == normalize_text(card_str):
            match_tracker[card_str].append(line)

# Check for validation failures
validation_passed = True
for card_str, matched_nodes in match_tracker.items():
    match_count = len(matched_nodes)
    card_num = card_alignment[card_str]
    
    if match_count == 0:
        print(f"❌ MISMATCH: Card {card_num} ('{card_str}') matched 0 lines in the XML.")
        validation_passed = False
    elif match_count > 1:
        print(f"❌ AMBIGUITY: Card {card_num} ('{card_str}') matched {match_count} different lines in the XML.")
        for i, node in enumerate(matched_nodes):
            print(f"    -> Match {i+1}: Line n=\"{node.get('n')}\": {''.join(node.itertext())}")
        validation_passed = False

# 3. Second Pass: Structural Insertion (Only runs if 1:1 match is guaranteed)
if validation_passed:
    print("✅ Validation Passed: Perfect one-to-one match found for all cards. Inserting milestones...")
    
    for card_str, matched_nodes in match_tracker.items():
        card_num = card_alignment[card_str]
        line_node = matched_nodes[0] # Guaranteed to be exactly one node
        
        # Create the milestone element
        milestone = etree.Element("milestone", unit="card", edRef="Storr", n=str(card_num))
        
        parent = line_node.getparent()
        # If the line is the first child element under <sp> (skipping <speaker>),
        # place the milestone immediately before the <sp> tag to match Storr's edition structure.
        if parent.tag == "{http://www.tei-c.org/ns/1.0}sp":
            children = list(parent)
            if children.index(line_node) == 1 or children.index(line_node) == 0:
                grandparent = parent.getparent()
                grandparent.insert(grandparent.index(parent), milestone)
                continue
                
        # Default: Insert right before the line element
        parent.insert(parent.index(line_node), milestone)

    # Save the properly structured output
    with open(outfname, "wb") as f:
        f.write(etree.tostring(tree, encoding="UTF-8", xml_declaration=True, pretty_print=False))
    print("🎉 File successfully saved to '"+outfname+"'")
else:
    print("⛔ Execution halted. Please resolve the string mismatches listed above before modifying the XML structure.")
```
