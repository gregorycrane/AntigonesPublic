```python
import re

fname = 'tlg0011.tlg002.boeckh1884-deu1.xml'

outfname = 'goodlines.' + fname
f = open(fname)
outf = open(outfname,'w')
curlnum = 0

for l in f:
    if(re.search('<l [^>]+>',l)):
        curlnum = curlnum + 1
        l = re.sub('<l [^>]+>','<l n="'+str(curlnum)+'">',l)
    print(l,end='',file=outf)
f.close()
outf.close()
```
