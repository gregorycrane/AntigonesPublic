```python
import re

basecards = {}
newcards = {}

f = open('/Users/gcrane/github/canonical-greekLit/data/tlg0011/tlg002/tlg0011.tlg002.perseus-grc2.xml')
for l in f:
    m = re.search('<milestone unit="card" n="([^<]+)"',l)
    if(m):
        basecards[m[1]] = 1
f.close()

f = open('/Users/gcrane/github/AntigonesPublic/data/tlg0011/tlg002/tlg0011.tlg002.hoelderlin1805-deu1.xml')
outf = open('/Users/gcrane/github/AntigonesPublic/staging/tlg0011.tlg002.hoelderlin1805-deu1.xml','w')
i = 0
for l in f:
    m = re.search('<milestone unit="card"[^<]+n="([^<]+)"',l)
    if(m):
        if(not m[1] in basecards):
            l = re.sub('<milestone unit="card"[^<]+n="([^<]+)"[^>]*>','',l,1)
            #print('skipping',m[1])
        else:
            i = i + 1
            #print(i,'keep',m[1])
            newcards[m[1]] = 1
    print(l,end='',file=outf)
f.close()

for foo in basecards:
    if( not foo in newcards):
        print('skipped',foo)
    

```
